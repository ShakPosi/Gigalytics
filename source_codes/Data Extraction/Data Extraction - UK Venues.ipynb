```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
import time
API_KEY = 'HPeybaQQsB9TAlBW'
ukEvents = pd.read_csv("ukEvents2.csv")
venueID = ukEvents.venueId.unique()
```

```python
venueID = sorted(venueID)
newVenueID = []
```

```python
count = 0
for i in venueID:
    try:
        print(int(i))
        newVenueID.append(int(i))
        print(count)
        count = count + 1
    except:
        continue
    
```

```python
ukVenuesIdcount = 0
venueAttributes = ['venueID', 'venueName', 'venueStreet', 'venuePostcode' , 'venueCity', 'venueCountry', 'venueCapacity', 'venueWebsite', 'venuePhone']
venueDataArray = []

for i in newVenueID:
    print(i)
    print(ukVenuesIdcount)
    ukVenuesIdcount = ukVenuesIdcount + 1
    try:
        eventsVenueRequest = requests.get('https://api.songkick.com/api/3.0/venues/' + str(i)  + '.json?apikey=' + API_KEY)
    except:
        print("Connection refused by the server..")
        print("Let me sleep for 60 seconds")
        print("ZZzzzz...")
        time.sleep(60)
        print("Was a nice sleep, now let me continue...")
        eventsVenueRequest = requests.get('https://api.songkick.com/api/3.0/venues/' + str(i)  + '.json?apikey=' + API_KEY)
        
    try:
        jsonData = json.loads(eventsVenueRequest.text)
        df = json_normalize(jsonData['resultsPage']) 
        venueData = [df["results.venue.id"][0], df["results.venue.displayName"][0], df["results.venue.street"][0], df["results.venue.zip"][0], df["results.venue.city.displayName"][0], df["results.venue.city.country.displayName"][0], df["results.venue.capacity"][0], df["results.venue.website"][0], df["results.venue.phone"][0]]
        venueDataArray.append(venueData)
    except:
        continue
myVenues = pd.DataFrame(data = venueDataArray, columns = venueAttributes)
myVenues.to_csv('ukVenues.csv', index=False)
```
